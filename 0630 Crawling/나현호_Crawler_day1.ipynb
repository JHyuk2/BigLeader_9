```python
from selenium import webdriver
import time, os, sys #time은 .sleep Method
```

```python
def ts(x): #sleep Method 편의를 위한 함수 정의
    time.sleep(x)#x초 만큼 정지

driver = webdriver.Chrome()#webdriver와 chrome,  Selenium 버전에 따라 Chrome 안에 

ts(2); driver.get('http://www.riss.kr/index.do') #get(Url)에 접근
ts(2); driver.maximize_window()#창을 최대화 액션
```

```python
main = driver.window_handles  #window 핸들러 설정

for handle in main: 
    if handle != main[0]: #창이 존재한다면
        driver.switch_to.window(handle) 
        driver.close() #팝업 창이므로 종료
        
driver.switch_to.window(driver.window_handles[0])
```
