길동이를 추가하고 싶음

```python
df2 = pd.DataFrame(exam_data)
```

```python
df2.set_index('이름', inplace=True)
```

```python
df2
```

```python
df2.loc['길동']=[70, 80, 90, 60]
```

```python
df2
```

```python
# 행과 열 (컬럼과 인덱스가)을 전치하고 싶을 때
df2.T
```

```python
# 평균을 이용한 산술
df2
```

```python
df2['총점'] = df2.수학 + df2.영어 + df2.음악 + df2.체육
df2['평균'] = df2['총점']/4
```

```python
df2
```

```python
# 점수별 등급 주기
grade = []  # 빈 리스트 만들고

# 평균 성적 가져오기
for avg in df2.평균:
    if avg >=90:
        grade.append('A')
    elif avg >=80:
        grade.append('B')
    else:
        grade.append('C')
```

```python
df2['성적']=grade
```

```python
df2
```

```python
def scholarship(row):
    print(row)    #굳이 안 써도 됨 쓰면 등급 나옴
    if row == "A":
        return '장학생'
    else:
        return '비장학생'
    
df2.성적 = df2.성적.apply(scholarship)
df2
```

```python
l1 = [{'name': 'John', 'job': "teacher"}, 
{'name': 'Nate', 'job': "student"}, 
{'name': 'Fred', 'job': "developer"}]
l2 = [{'name': 'Ed', 'job': "dentist"}, 
{'name': 'Jack', 'job': "farmer"}, 
{'name': 'Ted', 'job': "designer"}]
df1 = pd.DataFrame(l1, columns = ['name', 'job']) 
df2 = pd.DataFrame(l2, columns = ['name', 'job'])
```

```python
df1
```

```python
df2
```

```python
result=df1.append(df2, ignore_index=True)
result
```

```python
df3=pd.concat([df1, df2], ignore_index=True)
```

```python
df4=pd.concat([df1, df2], axis=1)  # axis : 행을 기준으로 할 것인가 열을 기준으로 할 것인가
```

merge (합치기)

```python
l1 = [{'name': 'John', 'job': "teacher"}, 
    {'name': 'Nate', 'job': "student"}, 
    {'name': 'Fred', 'job': "developer"}]

l2 = [{'name': 'Ed', 'work': "dentist"}, 
    {'name': 'Nate', 'work': "farmer"}, 
    {'name': 'Ted', 'work': "designer"}]

df3 = pd.DataFrame(l1, columns = ['name', 'job']) 
df4 = pd.DataFrame(l2, columns = ['name', 'work'])
```

```python
df3
```

```python
df4
```

```python
df_inner = pd.merge(df3, df4)  # 교집합
```

```python
df_inner
```

```python
df_outter = pd.merge(df3, df4, how = 'outer', on = 'name' )  # 차집합
```

```python
df_outter
```

```python
df_left = pd.merge(df3, df4, how = 'left', on = 'name' )  # 왼쪽
```

```python
df_left
```

```python
df_right = pd.merge(df3, df4, how = 'right', on = 'name' )  # 오른쪽
```

```python
df_right
```

NAN 값을 생존자/사망자 별 평균으로 대체하기

```python
# 생존자 나이 평균
mean1 = df[df['Survived'] == 1]['Age'].mean()
# 사망자 나이 평균
mean0 = df[df['Survived'] == 0]['Age'].mean()

print(mean1, mean0)
```

```python
df.loc[df['Survived'] == 1, 'Age'] = df[df['Survived'] == 1]['Age'].fillna(mean1)
df.loc[df['Survived'] == 0, 'Age'] = df[df['Survived'] == 0]['Age'].fillna(mean0)
```

```python
%matplotlib inline

s = [1,2,3,4,5,6,5,4,3,2,1]

s1 = pd.Series(s)
s1.plot()
```

```python
import matplotlib.pyplot as plt
```

```python
p1=['a', 'b', 'c', 'd', 'e']   # 애를 x축(항목)으로 나타내기
s1 = [1,2,3,2,1]
s2 = [2,4,3,5,2]

plt.xlabel('Test')  # x축 제목
plt.ylabel('Value')  # y축 제목
plt.title('Chart')  #차트 제목
plt.plot(p1, s1, "*y")  # 여기 넣음 x,y 좌표  # ""안에 있는 건 모양/색깔 지정 (*은 모양/ y는 색깔)
plt.plot(p1, s2)
plt.show()
```

```python
# 막대그래프
p1=['a', 'b', 'c', 'd', 'e']   # 애를 x축(항목)으로 나타내기
s1 = [1,2,3,2,1]
s2 = [2,4,3,5,2]

plt.xlabel('Test')
plt.ylabel('Value')
plt.title('Chart')
plt.plot(p1, s1)  # 여기 넣음 x,y 좌표
plt.bar(p1, s2)
plt.legend(['test1','test2'])  # 범례 지정
plt.show()
```

반복문

```python
from tqdm import tqdm
import time
```

```python
driver = webdriver.Chrome()
driver.get("https://www.opinet.co.kr/searRgSelect.do")
```

```python
area = driver.find_element(By.XPATH,'//*[@id="SIDO_NM0"]')
```

```python
area.send_keys('서울')
```

```python
gu_list_raw = driver.find_element(By.XPATH, '//*[@id="SIGUNGU_NM0"]')
```

```python
gu_list_raw
```

```python
for gu in gu_list_raw:
    print(gu)
```

```python
gu_list = gu_list_raw.find_elements(By.TAG_NAME, 'option')
```

```python
gu_list
```

```python
gu_names = [option.get_attribute('value') for option in gu_list]
```

```python
gu_names.remove('')
```

```python
gu_names
```

```python
for gu in tqdm(gu_names):
    element = driver.find_element(By.ID, "SIGUNGU_NM0")
    element.send_keys(gu)
    time.sleep(3)
    element.sel_gu = driver.find_element(By.XPATH, '''//*[@id="searRgSelect"]/span''').click()
    time.sleep(2)
    element_get_excel = driver.find_element(By.XPATH, '''//*[@id="glopopd_excel"]/span''').click()
    time.sleep(2)
```
