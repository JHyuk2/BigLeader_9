## 2. pandas.DataFrame

### 알아도 쓸데없는 잡지식) 왜 DataFrame 인가?

- Pandas를 보면, CapWords convention(카멜케이스)과 Snake case(언더바_를 이용한 네이밍 구분) 두 가지를 혼용하고 있다.
- pd.DataFrame : CamelCase, pd.data_range : snake case
- 위에서 DataFrame의 경우 class이기 때문에 카멜케이스로 사용되고, data_range는 function이기 때문에 snake case가 된다.
- [PEP8](https://peps.python.org/pep-0008/#class-names) 의 Python style guide 참조를 해보자.


```python
temp_dict = {'c0': [1,2,3], 'c1': [1,2,3], 'c2': [1,2,3], 'c3': [1,2,3], 'c4': [1,2,3]}
```

```python
df = pd.DataFrame(temp_dict)
```

```python
pp(df)
```

```python
pp(df.columns)
pp(df.index)
```

```python
df = pd.DataFrame([[15, '남', '덕영중'], [17, '여', '수리중']], index=['준서', '예은'], columns=['나이', '성별', '학교'])
df.head()
```

```python
# 방금 만든 df의 인덱스 : 준서, 예은
#                  컬럼 : 나이, 성별, 학교

# rename index/columns
rename_ins = ['학생1', '학생2']
rename_cols = ['연령', '남녀', '소속']

cols = df.columns
ins = df.index

rename_ins_dict = {}
rename_col_dict = {}

for col, rename_col in zip(cols, rename_cols):
    rename_col_dict[col] = rename_col
    
for index, rename_index in zip(ins, rename_ins):
    rename_ins_dict[index] = rename_index
    
pp(rename_col_dict)
pp(rename_ins_dict)
```

```python
df.rename(columns=rename_col_dict, index=rename_ins_dict)
```

```python
# rename만 했을 때, 원본에 변함은 없다.
df.head()
```

```python
df.rename(columns=rename_col_dict, index=rename_ins_dict, inplace=True) # inplace True라는 옵션을 주자, 원본에도 변화가 생겼다.
```

```python
df
```

```python
# df를 바꾸고 바로 적용하려면, 리스트 째로 집어넣으면 된다.
df.columns = rename_col
df.index = rename_index
df.head()
```

```python
grade = []

for avg in df_exam.평균:
    if avg >= 90:
        grade.append('A')
    elif avg>= 80:
        grade.append('B')
    else:
        grade.append('C')
        
pp(grade)
```

```python
# subplot을 사용하여 여러 개의 plot 을 한번에 찍을 수 있다.

s1 = [1, 2, 3, 4, 5,]
s2 = [2, 4, 3, 5, 2,]

plt.subplot(2, 2, 1) # 2 x 2, index=1 
plt.plot(s1, s2, marker='o')
plt.ylabel('Value')

plt.subplot(2, 2, 2) # 2 x 2, index=2
plt.bar(s1, s2, color='olive')
plt.title('Chart')

plt.subplot(2, 2, 3) # 2 x 2, index=3 
plt.plot(s1, s2, '*')
plt.xlabel('Test')

plt.subplot(2, 2, 4) # 2 x 2, index=4 
plt.plot(s1, s2, 'red')
plt.legend(['test1', 'test2'])
plt.show()
```
